```python

mean = combined_age_df.groupby('Year')['Civilian population aged 15-64 years Persons'].mean()
median =combined_age_df.groupby('Year')['Civilian population aged 15-64 years Persons'].median()
variance = combined_age_df.groupby('Year')['Civilian population aged 15-64 years Persons'].var()
stdv = combined_age_df.groupby('Year')['Civilian population aged 15-64 years Persons'].std()
sem = combined_age_df.groupby('Year')['Civilian population aged 15-64 years Persons'].sem()


# This method is the most straighforward, creating multiple series and putting them all together at the end.
summary_age_df = pd.DataFrame({"Mean": mean, "Median": median, "Variance": variance, "Standard Deviation": stdv, 
                          "SEM": sem})
summary_age_df
```
